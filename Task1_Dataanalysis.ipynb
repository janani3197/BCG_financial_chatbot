```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Task_1_data.xlsx')
```

```python
df
```

```python
df.info()
```

```python
# convert column Total Revenue (in millions) to int64
df['Total Revenue'] = df['Total Revenue'].astype('int64')
df['Year'] = df['Year'].astype(str)
```

```python
df
```

```python
# Calculate Year-over-Year Growth (%) and round to 2 decimal places
df["Revenue Growth (%)"] = df.groupby("Company")["Total Revenue"].pct_change().mul(100).round(2)
df["Net Income Growth (%)"] = df.groupby("Company")["Net Income"].pct_change().mul(100).round(2)
df["Assets Growth (%)"] = df.groupby("Company")["Total Assets"].pct_change().mul(100).round(2)
df["Liabilities Growth (%)"] = df.groupby("Company")["Total Liabilities"].pct_change().mul(100).round(2)
df["Cash Flow Growth (%)"] = df.groupby("Company")["Cash Flow from operating activities"].pct_change().mul(100).round(2)

# Display results
df
```

```python
#Dropping Nan values ( since task requires only past 3 years)
```

```python
df = df.dropna()
```

```python
df
```

```python
# Set plot style
sns.set(style="whitegrid")

# Plotting the Growth Rates for Revenue, Net Income, Assets, Liabilities, and Cash Flow
metrics = ['Revenue Growth (%)', 'Net Income Growth (%)', 'Assets Growth (%)', 
           'Liabilities Growth (%)', 'Cash Flow Growth (%)']

plt.figure(figsize=(12, 8))
for i, metric in enumerate(metrics):
    plt.subplot(2, 3, i+1)
    sns.lineplot(data=df, x='Year', y=metric, hue='Company', marker='o')
    plt.title(f'{metric} Over Time')
    plt.ylabel(f'{metric} (%)')
    plt.xlabel('Year')

plt.tight_layout()
plt.show()
```

```python
# Filter the data for 2024
df_2024 = df[df['Year'] == '2024']

# Plot a bar chart for the financial metrics in 2024
df_2024_metrics = df_2024[['Company', 'Total Revenue', 'Net Income', 'Total Assets', 
                            'Total Liabilities', 'Cash Flow from operating activities']]

df_2024_metrics.set_index('Company').T.plot(kind='bar', figsize=(10, 6))
plt.title('Comparison of Financial Metrics for 2024')
plt.ylabel('Amount in Millions')
plt.xlabel('Financial Metrics')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Trend Analysis Over Time
------------------------

Analyzing the trend of each financial metric over the 3 years can give insights into whether the company is improving or declining over time.

Line Plot showing trends for Revenue, Net Income, and other key financial metrics over the 3 years.

```python
# Heatmap for growth rates in 2024
df_2024_growth = df_2024[['Company', 'Revenue Growth (%)', 'Net Income Growth (%)', 
                          'Assets Growth (%)', 'Liabilities Growth (%)', 'Cash Flow Growth (%)']]

df_2024_growth.set_index('Company', inplace=True)

plt.figure(figsize=(10, 6))
sns.heatmap(df_2024_growth.T, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Growth Rate Comparison for 2024')
plt.ylabel('Financial Metrics')
plt.xlabel('Company')
plt.tight_layout()
plt.show()
```

Company Performance Summary
---------------------------

A combined table or chart summarizing overall performance, such as compound annual growth rate (CAGR) or simple year-over-year performance for each metric.

Calculated CAGR for each financial metric between 2022 and 2024 for each company.

```python
# Calculating CAGR for Total Revenue
df_cagr = df.pivot(index='Company', columns='Year', values='Total Revenue')
df_cagr['CAGR'] = ((df_cagr['2024'] / df_cagr['2022']) ** (1/2)) - 1
print(df_cagr[['CAGR']])
```

SUMMARY:
--------

Tesla is growing at a 9.51% CAGR, showing positive but more modest growth compared to Microsoft. However, this may be offset by the significant decline in Net Income Growth, which should be addressed.

Apple is facing challenges with a negative CAGR (-0.42%), indicating stagnation or decline over the observed period. Apple may need to focus on strategies to return to growth.

INSIGHTS:
---------
Microsoft is in a good position for sustained growth and is likely a stronger performer in the long term.

Tesla has room for improvement, especially in terms of profitability and Net Income, but it is still growing.

Apple needs to address the underlying causes of its negative growth to remain competitive and regain momentum.

Correlation Heatmap showing the correlation between financial metrics
---------------------------------------------------------------------

Performed a correlation analysis to see how closely related the different financial metrics are (e.g., Total Revenue vs. Net Income).

```python
# Calculate correlation matrix for financial metrics
correlation_matrix = df[['Total Revenue', 'Net Income', 'Total Assets', 
                                        'Total Liabilities', 'Cash Flow from operating activities']].corr()

# Plot heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix of Financial Metrics')
plt.tight_layout()
plt.show()
```

INSIGHTS: 
---------
1. There is a strong positive correlation across the majority of financial metrics, particularly between Net Income, Total Liabilities, and Cash Flow from Operating Activities. This suggests that companies with higher revenues, profitability, and asset bases tend to rely on liabilities for growth and also generate strong operational cash flow.

2. Total Revenue and Net Income are tightly linked, which is expected since more revenue often leads to higher profits.

3. Total Liabilities and Cash Flow from Operating Activities are highly correlated, which could indicate that companies with higher debt are successfully using it to finance profitable operations and generate cash.

IMPLICATIONS:
-------------
1. Companies that show high Total Revenue and Net Income may benefit from using debt (liabilities) to fuel growth without jeopardizing profitability.

2. Strong Cash Flow from Operating Activities is a good indicator of financial health, and it’s positively correlated with both Net Income and Total Liabilities.

3. If a company is increasing its Total Liabilities, it's important to ensure that this is being used to enhance Total Assets and Cash Flow, not just for short-term financial gains.
